## Loading the dataset

```python
#importing all the datasets for the 2019 cogo trip data
godata1 = pd.read_csv('201901-cogo-tripdata.csv')
godata2 = pd.read_csv('201902-cogo-tripdata.csv')
godata3 = pd.read_csv('201903-cogo-tripdata.csv')
godata4 = pd.read_csv('201904-cogo-tripdata.csv')
godata5 = pd.read_csv('201905-cogo-tripdata.csv')
godata6 = pd.read_csv('201906-cogo-tripdata')
godata7 = pd.read_csv('201907-cogo-tripdata')
godata8 = pd.read_csv('201908-cogo-tripdata')
godata9 = pd.read_csv('201909-cogo-tripdata.csv')
godata10 = pd.read_csv('201910-cogo-tripdata.csv')
godata11 = pd.read_csv('201911-cogo-tripdata.csv')
godata12 = pd.read_csv('201912-cogo-tripdata.csv')
```

```python
#defining a function to clean the datasets
def cleangodata(df):
    """This function will provide the detail information about the dataset, check for missing variables and then drop them,
    and finally check for duplicates."""
    print('Summary Statistics: \n')
    print('*'*70)
    print(df.info())
    print('*'*70)
    print('Missing Values: {}'.format(df.isna().sum()))
    print('*'*70)
    print('Missing Values Dropped: {}'.format(df.dropna(inplace= True)))
    print('*'*70)
    print('Duplicate Values: {}'.format(df.duplicated().sum()))
```

```python
#combining all these dataset into one master data
godata = pd.concat([godata1, godata2, godata3, godata4, godata5, godata6, godata7, godata8, godata9, godata10,
                    godata11, godata12], axis=0)
```

```python
godata.shape
```

```python
#changing the tripduration variable into integer datatype
godata['tripduration'] = godata.tripduration.str.replace(',', '').astype(float)
godata['tripduration'] = godata.tripduration.astype(int)

#changing the birthyear variable into integer datatype
godata['birthyear'] = godata.birthyear.astype('int')
```

```python
godata.info()
```

```python
#changing the start_date and start_time from object to datetime
godata['start_date'] = pd.to_datetime(godata['start_time'])
godata['start_time'] = pd.to_datetime(godata['start_time'])


#changing the end_date and end_time from object to datetime
godata['end_date'] = pd.to_datetime(godata['end_time'])
godata['end_time'] = pd.to_datetime(godata['end_time'])
```

```python
#dropping redundant variables
godata.drop(['bikeid','from_station_location', 'from_station_id','to_station_location','to_station_id'], 
            axis=1, inplace=True)  
```

## Univariate Exploration

```python
godata.tripduration.describe() #checking for the min and max values
```

```python
# finding out the distribution of the tripduration
binsize = 1000
bins = np.arange(0, godata['tripduration'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = godata, x = 'tripduration', bins = bins)
plt.xlabel('Trip duration (Seconds)')
plt.show()
```

>The data is skewed to the right because of an outlier in the tripduration feature. Therefore I will scale it with the log 
scale in order to transform it into a normal distribution

```python
np.log10(godata['tripduration'].describe()) #checking for the min and max values after scaling
```

```python
#scaling the distribution of the tripduration feature
bins = 10 ** np.arange(0.1, 5+0.1, 0.1)
ticks = [1, 2, 5, 10, 50, 100, 1000, 10000, 100000]
labels = ['{}'.format(val) for val in ticks]

plt.figure(figsize=[8, 5])
plt.hist(data = godata, x = 'tripduration', bins = bins)
plt.xscale('log')
plt.xticks(ticks, labels)
plt.xlabel('Trip duration (Seconds)');
```

>After scaling, it can be seen that most of the trip duration fell between 100 and 1000 seconds while few others fell 
below and above these values.

Next: Plot a pie chart for the usertype variable

```python
#Which of the usertype dorminates?
sorted_counts = godata['usertype'].value_counts()
type_label = ['Subscriber', 'Customer']
base_color = [sb.color_palette()[0], sb.color_palette()[3]]
#ploting the usertpe on a pie chart
plt.pie(sorted_counts, labels = type_label, startangle = 90, counterclock = False, colors = base_color, autopct='%1.1f%%', 
       pctdistance=0.9)
plt.axis('square');
```

```python
#splitting the start_date into year, month and day
godata['start_year'] = godata['start_date'].dt.year
godata['start_month'] = godata['start_date'].dt.month_name()
godata['start_day'] = godata['start_date'].dt.day_name()
```

```python
#creating the age variable from the start_year and birthyear
godata['age'] = godata['start_year'] - godata['birthyear']
```

Plotting a histogram for the age variable

```python
bins = np.arange(20, godata['age'].max()+5, 5)
plt.hist(data = godata, x = 'age', bins = bins)
plt.xlabel('ages');
```

Plotting a bar chart for the start_month variable

>The month with the most bicycle riders is May. Considering the fact that May is the last month of the spring season in America, and this season encourages more outdoor activities, it could be one of the reasons why there is so much rides in this month.

Plotting a bar chart for the start_day variable

## Bivariate Exploration

Plotting a scatter plot to find the relationship between tripduration and birthyear

```python
#is there any relationship between the birth year of riders and their trip durations?
# Scatter plot
plt.figure(figsize = [20, 5])
#the first plot shows the scatter plot between birthyear and tripduration
plt.subplot(1, 2, 1)
sb.regplot(data = godata, x = 'birthyear', y = 'tripduration')
plt.xlabel('Birth Year')
plt.ylabel('Trip Duration');

#the second plot shows the same scatter plot but with some transformations using jittering and transparency options
plt.subplot(1, 2, 2)
sb.regplot(data = godata, x = 'birthyear', y = 'tripduration', truncate=False, x_jitter=0.3, scatter_kws={'alpha':1/20});
```

>From the visualizations above, it can be seen that there is no clear relationship between birth year of riders and their trip durations.

## Multivariate Exploration

Plotting a scatter plot for tripduration, age and usertype

```python
g = sb.FacetGrid(data = godata, hue = 'usertype', height = 5,
                 palette = 'colorblind')
g.map(plt.scatter, 'age', 'tripduration')
g.add_legend();
```

There seem to be no significant relationship between age and tripduration. However, it can be seen that 'customer' usertype tend to have lower trip duration compared to the 'subscribers'.

Plotting a boxplot for start_month, age and gender

```python
g = sb.FacetGrid(data = godata, col = 'gender', height = 10)
g.map(sb.boxplot, 'start_month', 'age')
# plt.xticks(rotation=30);
```

There is greater variability in the distribution of the male ages across the the entire months compared to that of female. Also, the age outiers are larger for male compared to female. 

Plotting a pointplot for start_month, age and gender

```python
fig = plt.figure(figsize = [8,6])
ax = sb.pointplot(data = godata, x = 'start_month', y = 'age', hue = 'usertype',
           palette = 'Blues', linestyles = '', dodge = 0.4)
plt.title('Interactions of CoGo Riders across Usertype and Startmonth base on their Ages')
plt.xticks(rotation=30)
plt.show();
```

The majority of the 'subscribers' seem to be in their 40's. The larger error bars shows the fewer 'customers' during each month of the year.
The majority of the 'customers' seem to be concentrated around May and June and their ages are below 40 years old.
